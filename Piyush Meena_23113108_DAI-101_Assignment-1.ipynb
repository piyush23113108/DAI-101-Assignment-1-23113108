### **DATA CLEANING**

Data cleaning is the process of identifying, correcting, or removing errors and inconsistencies in data to ensure its quality, reliability, and usability for analysis.

*METHODS ARE Z-SCORE,INTERQUARTILE RANGE,REMOVE OUTLIERS OR REPLACE THEM QITH MEAN MEDIAN AND MODE*

```python
import scipy.stats
# Check for outliers
# Option 1: Z-Score
z = np.abs(scipy.stats.zscore(df["release_year"]))
print(z)

# Option 2: Interquartile Range (IQR)
Q1 = df["release_year"].quantile(0.25)
Q3 = df["release_year"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(df[(df["release_year"] < lower) | (df["release_year"] > upper)])

# Handle outliers
# Option 1: Drop outliers
df = df[(z < 3)]

# Option 2: Replace outliers with mean/median/mode
df["release_year"][df["release_year"] < lower] = lower
df["release_year"][df["release_year"] > upper] = upper
```

By identifying and handling outliers, the analysis improves the robustness and accuracy of subsequent statistical analyses or machine learning models. Dropping outliers may simplify the dataset but can also lead to the loss of valuable information. Conversely, replacing outliers can help retain data while minimizing their impact on results. The choice of method depends on the specific context and goals of the analysis.

## **Statistical Tests**

# -**ANNOVA**

```python
from scipy.stats import f_oneway
groups = [group['le_type'].dropna() for name, group in df.groupby('release_year')]
f_stat, p_val = f_oneway(*groups)

print(f"F-Statistic: {f_stat}")
print(f"P-value: {p_val}")
```

### **Conclusion**

If the p-value is less than 0.05, you can conclude that there are statistically significant differences in le_type values among the different release years, indicating that the release year may have an effect on the categorical variable represented by le_type. If the p-value is greater than 0.05, it suggests no significant differences exist among the means of the groups.

# **CHI2 TEST**

```python
# Selecting categorical columns based on the first 20 rows
cat_cols = [col for col in df.columns if df.iloc[:20][col].dtypes == "object"]

# Selecting numerical columns (int64 or float64) based on the first 20 rows
num_cols = [col for col in df.columns if df.iloc[:20][col].dtypes == "int64" or df.iloc[:20][col].dtypes == "float64"]

cat_cols, num_cols
```

```python
for col1, col2 in combinations(cat_cols, 2):
    # Create contingency table
    contingency_table = pd.crosstab(df[col1], df[col2])
    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi-Square Test between {col1} and {col2}:")
    print(f"Chi2 Statistic: {chi2:.4f}, p-value: {p:.4f}, Degrees of Freedom: {dof}")
```

This code conducts pairwise Chi-Square tests to assess the association between categorical variables within a DataFrame. By generating all possible combinations of two categorical columns, it constructs a contingency table for each pair, summarizing the frequency of occurrences for every combination of values. The Chi-Square test is then applied to each table, producing key statistical outputs, including the Chi-Square statistic, p-value, and degrees of freedom. These results indicate whether the relationship between the two categorical variables is statistically significant, helping to identify potential dependencies or correlations. A low p-value suggests a strong association, while a high p-value indicates independence between the variables.
